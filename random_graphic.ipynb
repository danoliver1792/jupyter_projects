```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
# Gerando 100 pontos de dados aleatórios em 3 dimensões
x, y, scale = np.random.randn(3, 100)
fig, ax = plt.subplots()

# Mapeando cada um em um gráfico de dispersão com a biblioteca Matplotlib
ax.scatter(x=x, y=y, c=scale, s=np.abs(scale) * 500)
ax.set(title='Alguns dados aleatórios criados com JupyterLab')
plt.show()
```
