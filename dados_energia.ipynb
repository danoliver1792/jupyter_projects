```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
# Lendo o arquivo T1.csv
turbina = pd.read_csv('T1.csv')

# Mudando o nome das colunas
turbina.columns = ['Data/Hora', 'Potencia (KW)', 'Velocidade do Vento (m/s)', 'Curva Teorica (KWh)', 'Direcao do Vento (°)']
# Vamos deletar a coluna 'Direcao do Vento (°)'
del turbina['Direcao do Vento (°)']
# Transformando o tipo da coluna 'Data/Hora' em datetime
turbina['Data/Hora'] = pd.to_datetime(turbina['Data/Hora'])

display(turbina)
```

```python
# Plotando um gráfico de dispersão da coluna 'Velocidade do Vento' com 'Potencia'
sns.scatterplot(data=turbina, x='Velocidade do Vento (m/s)', y='Potencia (KW)')
```

O gráfico acima corresponde que a cada metro por segundo a turbina girou em uma determinada potência.

```python
# Plotando a curva teórica 
sns.scatterplot(data=turbina, x='Velocidade do Vento (m/s)', y='Curva Teorica (KWh)')
```

Curva Teórica - Quanto menos vento, menos energia será produzida. Quando aumenta o vento, a potência aumenta até chegar ao pico máximo.

Ex.:
Para 10 m/s a turbina deveria está girando 2500 KWh
Para 20 m/s a turbina deveria está girando 3500 KWh

```python
# Criando limites aceitáveis
pot_real = turbina['Potencia (KW)'].tolist()
pot_teorica = turbina['Curva Teorica (KWh)'].tolist()

# Criando lista de potência máxima (valor da turbina teórica + 5%) e potência mínima (potência teórica - 5%)
pot_max = []
pot_min = []
dentro_limite = []

for potencia in pot_teorica:
    pot_max.append(potencia * 1.05)
    pot_min.append(potencia * 0.95)

# Comparando as potências e vendo se está dentro ou fora da potência máxima
for p, potencia in enumerate(pot_real):
    if potencia >= pot_min[p] and potencia <= pot_max[p]:
        dentro_limite.append('Dentro')
    elif potencia == 0:
        dentro_limite.append('Zero')
    else:
        dentro_limite.append('Fora')
        
# Verificando a porcentagem de pontos que está dentro do nosso limite
print(dentro_limite.count('Dentro')/len(dentro_limite) * 100, '%')
```

Com limites aceitáveis, podemos dizer que é a margem de erro com 5% para cima ou 5% para baixo
5% para cima significa uma supereficiência fora do comum e 5% para baixo significa que pode está com problema
